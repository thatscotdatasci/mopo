```python
scores_2_epochs_df = get_scores_df(experiments_2_epochs, evaluation_datasets)
scores_4_epochs_df = get_scores_df(experiments_4_epochs, evaluation_datasets)
scores_11_epochs_df = get_scores_df(experiments_11_epochs, evaluation_datasets)
```

```python
scores_2_epochs_groups = (
    scores_2_epochs_df.groupby(['training_dataset', 'rex', 'rex_beta', 'evaluation_dataset'])
)
scores_2_epochs_avgs_df = scores_2_epochs_groups.mean().reset_index()
scores_2_epochs_mins_df = scores_2_epochs_groups.min().reset_index()
scores_2_epochs_maxs_df = scores_2_epochs_groups.max().reset_index()

scores_4_epochs_groups = (
    scores_4_epochs_df.groupby(['training_dataset', 'rex', 'rex_beta', 'evaluation_dataset'])
)
scores_4_epochs_avgs_df = scores_4_epochs_groups.mean().reset_index()
scores_4_epochs_mins_df = scores_4_epochs_groups.min().reset_index()
scores_4_epochs_maxs_df = scores_4_epochs_groups.max().reset_index()

scores_11_epochs_groups = (
    scores_11_epochs_df.groupby(['training_dataset', 'rex', 'rex_beta', 'evaluation_dataset'])
)
scores_11_epochs_avgs_df = scores_11_epochs_groups.mean().reset_index()
scores_11_epochs_mins_df = scores_11_epochs_groups.min().reset_index()
scores_11_epochs_maxs_df = scores_11_epochs_groups.max().reset_index()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(30, 10))

lss =  ['-', '--']
labels = ['2 Epochs', '4 Epochs', '11 Epochs']

metric = 'overall_mse'
for i, (exp_avgs, exp_mins, exp_maxs) in enumerate([
    (scores_2_epochs_avgs_df, scores_2_epochs_mins_df, scores_2_epochs_maxs_df),
    (scores_4_epochs_avgs_df, scores_4_epochs_mins_df, scores_4_epochs_maxs_df),
    (scores_11_epochs_avgs_df, scores_11_epochs_mins_df, scores_11_epochs_maxs_df),
]):
    plot_avgs = exp_avgs.set_index('evaluation_dataset').loc[evaluation_datasets][metric]
    plot_mins = exp_mins.set_index('evaluation_dataset').loc[evaluation_datasets][metric]
    plot_maxs = exp_maxs.set_index('evaluation_dataset').loc[evaluation_datasets][metric]
    ax.plot(evaluation_datasets, plot_avgs, label=labels[i], c=cols[i])
    ax.fill_between(evaluation_datasets, plot_mins, plot_maxs, alpha=0.25, color=cols[i])
ax.set_xticks(ax.get_xticks(), evaluation_datasets, rotation=80)
ax.set_xlabel('Evaluation Dataset')
ax.set_ylabel('Overall MSE')
ax.legend()
```
