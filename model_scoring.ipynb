```python
plt.rc('font', size=16)
```

```python
def plot_score(results_dicts, metric, title):
    fig, ax = plt.subplots(1, 1, figsize=(10,10))

    if len(results_dicts) == 2:
        labels = ["                            All Data", ""]
    elif len(results_dicts) == 6:
        labels = ["                            All Data", "",  "                            Training Data", " ", "                            Heldout Data", "  "]
    
    barlist = ax.bar(
        labels,
        [i[metric] for i in results_dicts]
    )

    for i in range(0, len(results_dicts), 2):
        barlist[i].set_color('#1f77b4')
    barlist[0].set_label('No REx')

    for i in range(1, len(results_dicts), 2):
        barlist[i].set_color('#ff7f0e')
    barlist[1].set_label('REx')

    ax.set_title(title)
    ax.set_ylabel(metric)
    # ax.set_ylim(0, 60)
    ax.legend()
    # ax.tick_params(axis='x', labelrotation=45)

    patches = ax.patches
    for rect in patches:
        label = f'{rect.get_height():.4f}'
        ax.text(
            rect.get_x() + rect.get_width() / 2, rect.get_height(), label, ha="center", va="bottom", fontdict={'size': 16}
        )
```

# D3RLPY-PAP5

```python
d3rlpy_pap5_core_datasets = [
    "D3RLPY-PAP5-P0-3",
    "D3RLPY-PAP5-P1-4",
    "D3RLPY-PAP5-P0_100000",
    "D3RLPY-PAP5-P4_100000",
    "D3RLPY-PAP5_100000"
]

d3rlpy_pap5_training_datasets = [
    "D3RLPY-PAP5-P0_25000",
    "D3RLPY-PAP5-P1_25000",
    "D3RLPY-PAP5-P2_25000",
    "D3RLPY-PAP5-P3_25000",
    "D3RLPY-PAP5-P4_25000",
]

d3rlpy_pap5_pol_datasets = [
    "D3RLPY-PAP5-P0_100000",
    "D3RLPY-PAP5-P1_100000",
    "D3RLPY-PAP5-P2_100000",
    "D3RLPY-PAP5-P3_100000",
    "D3RLPY-PAP5-P4_100000",
]

d3rlpy_pap5_datasets = [
    "D3RLPY-PAP5",
    "D3RLPY-PAP5-P0-3",
    "D3RLPY-PAP5-P1-4",
    "D3RLPY-PAP5_100000",
    "D3RLPY-PAP5-P0_25000",
    "D3RLPY-PAP5-P1_25000",
    "D3RLPY-PAP5-P2_25000",
    "D3RLPY-PAP5-P3_25000",
    "D3RLPY-PAP5-P4_25000",
    "D3RLPY-PAP5-P0_100000",
    "D3RLPY-PAP5-P1_100000",
    "D3RLPY-PAP5-P2_100000",
    "D3RLPY-PAP5-P3_100000",
    "D3RLPY-PAP5-P4_100000",
]

d3rlpy_pap5_experiments = ["MP100", "MP101", "MP102", "MP103", "MP104", "MP105", "MP106", "MP107", "MP108", "MP109", "MP110", "MP111", "MP124", "MP125", "MP126", "MP134", "MP135", "MP136", "MP137", "MP138", "MP139"]#, "MP130", "MP131", "MP132"]
```

```python
d3rlpy_pap5_scores = {}
for exp in d3rlpy_pap5_experiments:
    exp_details = get_experiment_details(exp)
    key = ' - '.join((exp_details.name, exp_details.dataset, f'REx: {exp_details.rex}'))
    for data in d3rlpy_pap5_datasets:
        score = get_score(exp, data)
        score['rex'] = exp_details.rex
        score['rex_beta'] = exp_details.rex_beta or 0.
        score['seed'] = exp_details.seed
        score['training_dataset'] = exp_details.dataset
        d3rlpy_pap5_scores[(key, data)] = score
d3rlpy_pap5_scores_df = (
    pd.DataFrame().from_dict(d3rlpy_pap5_scores, orient='index')[['training_dataset', 'rex', 'rex_beta', 'seed', 'observation_mse', 'reward_mse', 'log_prob', 'next_obs_log_prob', 'reward_log_prob']].
    reset_index().rename(columns={'level_0': 'experiment', 'level_1': 'evaluation_dataset'})
)
```

```python
d3rlpy_pap5_scores_groups = (
    d3rlpy_pap5_scores_df.groupby(['training_dataset', 'rex', 'rex_beta', 'evaluation_dataset'])
)
d3rlpy_pap5_scores_avgs_df = d3rlpy_pap5_scores_groups.mean().reset_index()
d3rlpy_pap5_scores_stds_df = d3rlpy_pap5_scores_groups.std().reset_index().replace(np.NaN, 0)
```
