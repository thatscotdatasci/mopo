```python
from collections import namedtuple

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from dogo.results import get_sac_pools, get_experiment_details, PoolArrs
```

```python
plt.rc('font', size=10)
cols = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
```

```python
EXP_LIST = [
    "HO265", # No REx - MOPO Pen 5 - Rollout Length 10 - MIXED-RT-1
    "HO266",
    "HO267",
    "HO274", # REx 10 - MOPO Pen 5 - Rollout Length 10 - MIXED-RT-1
    "HO275",
    "HO276",
    "HO313", # No REx - MOPO Pen 5 - Rollout Length 10 - MIXED-RT-4
    "HO314",
    "HO315",
    "HO322", # REx 10 - MOPO Pen 5 - Rollout Length 10 - MIXED-RT-4
    "HO323",
    "HO324",
    "HO217", # No REx - MOPO Pen 5 - Rollout Length 10 - MIXED-RT-5
    "HO218",
    "HO219",
    "HO226", # REx 10 - MOPO Pen 5 - Rollout Length 10 - MIXED-RT-5
    "HO227",
    "HO228",
]
SUBSAMPLE_SIZE = None
training_results = {}
for exp in EXP_LIST:
    # print(exp)
    training_results[exp] = PoolArrs(*get_sac_pools(exp, subsample_size=SUBSAMPLE_SIZE))
```

```python
fig, ax = plt.subplots(1, 1, figsize=(20,10))

labels = []
rmse_arrs = []
for i, (exp, rex, beta) in enumerate([
    ("HO265", False, 0.0),
    ("HO266", False, 0.0),
    ("HO267", False, 0.0),
    ("HO274", True, 20.0),
    ("HO275", True, 20.0),
    ("HO276", True, 20.0),
]):
    rmse_vals = np.sqrt(training_results[exp].mse_results).flatten()
    labels.append(f'{exp} - REx: {rex} - Beta: {beta}\nMean: {np.nanmean(rmse_vals):.2f}')

    rmse_vals = np.nan_to_num(rmse_vals, nan=1e6)
    rmse_arrs.append(rmse_vals)

ax.boxplot(rmse_arrs, showmeans=False, showfliers=False, labels=labels)

ax.set_title('H-MIXED-RT-1 - Mixed - MOPO Penalty: 5.0 - Rollout-Length: 10')
ax.set_xlabel('Experiment')
ax.set_ylabel('Overall RMSE')
# ax.set_ylim(-0.01,0.35)
```
