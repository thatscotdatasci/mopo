```python
SEEDS = [1443, 1234, 4321, 144, 244, 344]
```

```python
def get_policy_stochastic_results(policy_exp_name: str):
    policy_exp = get_experiment_details(policy_exp_name)
    dynamics_exp_name = policy_exp.dynamics_model_exp
    fake_rewards = []
    fake_state_action = []
    for seed in SEEDS:
        fake_rewards_seed = np.load(os.path.expanduser(f"~/rds/rds-dsk-lab-eWkDxBhxBrQ/dimorl/code/dogo_results/mopo/analysis/policy/{dynamics_exp_name}_{policy_exp_name}_dmFalse_dpTrue_{seed}_fake_unpen_rewards.npy"))
        fake_state_action_seed = np.load(os.path.expanduser(f"~/rds/rds-dsk-lab-eWkDxBhxBrQ/dimorl/code/dogo_results/mopo/analysis/policy/{dynamics_exp_name}_{policy_exp_name}_dmFalse_dpTrue_{seed}_fake_state_action.npy"))

        fake_state_action_seed[:,:,1:8] = np.arctan2(np.sin(fake_state_action_seed[:,:,1:8]), np.cos(fake_state_action_seed[:,:,1:8]))

        fake_rewards.append(fake_rewards_seed)
        fake_state_action.append(fake_state_action_seed)
    
    max_len = max(i.shape[1] for i in fake_rewards)

    fake_rewards = np.vstack([np.pad(i, ((0,0),(0,max_len-i.shape[1]),(0,0)), mode='constant', constant_values=0) for i in fake_rewards])
    fake_state_action = np.vstack([np.pad(i, ((0,0),(0,max_len-i.shape[1]),(0,0)), mode='constant', constant_values=0) for i in fake_state_action])

    return fake_rewards, fake_state_action
```

```python
def get_policy_stochistic_nan_counts(policy_exp_label_groups):
    all_labels = []
    nan_counts = np.zeros(len(policy_exp_label_groups))
    for i, (policy_exp_set, label) in enumerate(policy_exp_label_groups):
        for policy_exp in policy_exp_set:
            rewards, _ = get_policy_stochastic_results(policy_exp)
            nan_counts[i] += np.isnan(rewards.sum(axis=1).flatten()).sum()
        all_labels.append(label)
    return nan_counts, all_labels
```
