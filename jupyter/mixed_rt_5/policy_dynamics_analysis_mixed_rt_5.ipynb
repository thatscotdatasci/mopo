```python
SEEDS = [1443, 1234, 4321, 144, 244, 344]
```

```python
def get_policy_stochastic_results(policy_exp_name: str):
    policy_exp = get_experiment_details(policy_exp_name)
    dynamics_exp_name = policy_exp.dynamics_model_exp
    fake_rewards = None
    fake_state_action = None
    for seed in SEEDS:
        fake_rewards_seed = np.load(os.path.expanduser(f"~/rds/rds-dsk-lab-eWkDxBhxBrQ/dimorl/code/dogo_results/mopo/analysis/policy/{dynamics_exp_name}_{policy_exp_name}_dmFalse_dpTrue_{seed}_fake_unpen_rewards.npy"))
        fake_state_action_seed = np.load(os.path.expanduser(f"~/rds/rds-dsk-lab-eWkDxBhxBrQ/dimorl/code/dogo_results/mopo/analysis/policy/{dynamics_exp_name}_{policy_exp_name}_dmFalse_dpTrue_{seed}_fake_state_action.npy"))

        fake_state_action_seed[:,:,1:8] = np.arctan2(np.sin(fake_state_action_seed[:,:,1:8]), np.cos(fake_state_action_seed[:,:,1:8]))

        if fake_rewards is None:
            fake_rewards = fake_rewards_seed
            fake_state_action = fake_state_action_seed
        else:
            fake_rewards = np.vstack((fake_rewards, fake_rewards_seed))
            fake_state_action = np.vstack((fake_state_action, fake_state_action_seed))
    return fake_rewards, fake_state_action
```

```python
def get_policy_stochistic_nan_counts(policy_exp_label_groups):
    all_labels = []
    nan_counts = np.zeros(len(policy_exp_label_groups))
    for i, (policy_exp_set, label) in enumerate(policy_exp_label_groups):
        for policy_exp in policy_exp_set:
            rewards, _ = get_policy_stochastic_results(policy_exp)
            nan_counts[i] += np.isnan(rewards.sum(axis=1).flatten()).sum()
        all_labels.append(label)
    return nan_counts, all_labels
```

```python
#Rollout length 5
no_rex_nan_count, no_rex_nan_labels = get_policy_stochistic_nan_counts([
    (("HC151", "HC152", "HC153"), 'No REx $\\lambda=0.0$'),
    (("HC154", "HC155", "HC156"), 'REx: $\\beta=5.0$ $\\lambda=0.0$'),
    (("HC157", "HC158", "HC159"), 'REx: $\\beta=10.0$ $\\lambda=0.0$'),
    (("HC160", "HC161", "HC162"), 'REx: $\\beta=20.0$ $\\lambda=0.0$'),
    (("HC175", "HC176", "HC177"), 'No REx $\\lambda=1.0$'),
    (("HC178", "HC179", "HC180"), 'REx: $\\beta=5.0$ $\\lambda=1.0$'),
    (("HC181", "HC182", "HC183"), 'REx: $\\beta=10.0$ $\\lambda=1.0$'),
    (("HC184", "HC185", "HC186"), 'REx: $\\beta=20.0$ $\\lambda=1.0$'),
    (("HC187", "HC188", "HC189"), 'No REx $\\lambda=5.0$'),
    (("HC190", "HC191", "HC192"), 'REx: $\\beta=5.0$ $\\lambda=5.0$'),
    (("HC193", "HC194", "HC195"), 'REx: $\\beta=10.0$ $\\lambda=5.0$'),
    (("HC196", "HC197", "HC198"), 'REx: $\\beta=20.0$ $\\lambda=5.0$'),
])

fig, ax = plt.subplots(1,1,figsize=(20,10))
ax.bar(no_rex_nan_labels, no_rex_nan_count)
ax.set_xlabel('REx and MOPO Penalty Coefficients')
ax.set_ylabel('Experiments with Extreme Reward Predictions')
ax.set_xticks(ax.get_xticks(), no_rex_nan_labels, rotation=30)
```
